```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# read in the CSV
df = pd.read_csv("./Resources/survey lung cancer.csv")
df.head()
```

```python
# plotting the smoking vs lung cancer
df.plot(x = "SMOKING", y = "LUNG_CANCER", kind = "scatter")
```

```python
# Plotting the histogram
df_SMOKING = df[df["SMOKING"] == 2]
df_n_SMOKING = df[df["SMOKING"] == 1]

plt.hist([df_SMOKING["AGE"], df_n_SMOKING["AGE"]], color = ["Black", "Red"], label = ["w_SMOKING", "wo_SMOKING"], bins = 50)
plt.xlabel("Age")
plt.title("Age Distribution")
plt.legend()

plt.show()
```

```python
# of the people who has lung cancer what is the distribution of age
df.boxplot(column=["AGE"], by="LUNG_CANCER")
```
